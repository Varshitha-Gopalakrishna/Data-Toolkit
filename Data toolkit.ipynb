```python
"""
Q3) Explain the following terms:
    1) The difference in np.array, np.asarray and np.asanyarray
    2) The difference between Deep copy and shallow copy
Answer-
    1) The difference in np.array, np.asarray and np.asanyarray -
       a) np.array - Creates a new NumPy array from an input object (like a list, tuple, or another array). By default, it always creates a 
                     new array (deep copy), even if the input is already a NumPy array, unless you specify copy=False.

       b) np.asarray - Converts the input to a NumPy array, but does not create a copy if the input is already a NumPy array
                       (it returns the original array).
        
       c) np.asanyarray - if the input is a subclass of NumPy arrays (like a matrix or masked_array), it returns the subclass. 
                          This is important when working with specialized array types.

    2) The difference between Deep copy and shallow copy -
       a) Deep copy - In a deep copy, both the object and its data are completely copied. Any changes made to the copied object do not 
                      affect the original. Both the new object and its data are fully independent of the original.
                      
       b) Shallow copy - In a shallow copy, a new object is created, but the elements (or references) within the object are still the same 
                         as those in the original object. The new object is created, but it shares the same data.
"""
```

```python
#Example of np.array
Arr = np.array([1, 2, 3])
Arr
```

```python
#Example of np.asarray
arr1 = np.asarray([1, 2, 3])
print(arr1)

#If the input is already a NumPy array, no copy is made
arr1 = np.array([1, 2, 3])
arr2 = np.asarray(arr1)  
print(arr2)
```

```python
#Example of np.asanyarray
arr = np.asanyarray([1, 2, 3])
print(arr)

#If the input is already a subclass of an array, no copy is made.
arr1 = np.matrix([1, 2, 3])
arr2 = np.asanyarray(arr1)  
print(arr1)
print(arr2)

# arr2 remains a matrix, not a regular array
```

```python
#Example of Shallow copy

a = [1, 2, 3]
b = [4, 5, 6]

a = b     

print(a)
print(b)

a[0] = 66         #if changes are made in any one of the list, both lists will be changed.

print(a)
print(b)

b[1] = 6

print(a)
print(b)
```

```python
#Example of deep copy

a = [1, 3, 5]
b = [2, 4, 6]

a = b.copy()
print(a)
print(b)

a[2] = 8

print(a)       #the changes made in a does not affect b as they both point to different memory blocks.
print(b) 
```

```python
"""
Q14) Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks 
     using Matplotlib:
     a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for
        'Humidity').
     b) Label the x-axis as 'Date'.
     c) Set the title of the plot as 'Temperature and Humidity Over Time'
Answer-
"""

#time-series dataset
np.random.seed(42)
dates = pd.date_range('2023-01-01', periods=100)
temperature = np.random.uniform(15, 30, size=100)
humidity = np.random.uniform(30, 90, size=100)

df = pd.DataFrame({
    'Date': dates,
    'Temperature': temperature,
    'Humidity': humidity
})

#a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
fig, ax1 = plt.subplots()

# Left y-axis for Temperature
ax1.plot(df['Date'], df['Temperature'], 'g-', label='Temperature')

#b) Label the x-axis as 'Date'.
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='g')

# Right y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], 'b-', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='b')

#c) Set the title of the plot as 'Temperature and Humidity Over Time'
plt.title('Temperature and Humidity Over Time')

# Show plot
plt.show()
```

```python
"""
Q17) Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a 
     legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
Answer-
"""

x = np.random.randn(100)
y = np.random.randn(100)

quadrants = []
for i in range(len(x)):
    if x[i] >= 0 and y[i] >= 0:
        quadrants.append('Quadrant 1')
    elif x[i] < 0 and y[i] >= 0:
        quadrants.append('Quadrant 2')
    elif x[i] < 0 and y[i] < 0:
        quadrants.append('Quadrant 3')
    else:
        quadrants.append('Quadrant 4')

#DataFrame for plotting
df = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

#scatter plot
sns.scatterplot(x='x', y='y', hue='Quadrant', data=df)

#title and labels
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend(title='Quadrant')
plt.show()
```

```python
"""
Q19) Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to 
     display exact values, label the axes, and set the title as 'Random Categorical Bar Chart.
Answer-
"""
```

```python
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Viridis256


categories = ['A', 'B', 'C', 'D', 'E', 'F']
values = np.random.randint(10, 100, size=6)

df = pd.DataFrame({'categories': categories, 'values': values})

source = ColumnDataSource(df)

p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Adding bar glyphs with color mapping based on values
p.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white', fill_color=factor_cmap('categories', palette=Viridis256, factors=categories))

# Adding hover tool to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

show(p)
```
